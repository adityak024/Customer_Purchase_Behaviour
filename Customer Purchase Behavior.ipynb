```python
# how to deal with the missing data, specially by mean/median
# mostly by median because mean is affected by outliers and median is robust to outliers.
# now we have to categorise each item by  reviews because reviews of different products do not overlap
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a new column

labels = ['Young Adult','Adult','Middle Aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df.columns
```

```python
df[['age','age_group']].head(10)
```

```python
#create a column for purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
#to check that does both columns carry similiar informatiom, if yes, then remove one column

(df['discount_applied'] == df['promo_code_used']).all()
```

```python
#so we will drop one column

df = df.drop('promo_code_used', axis=1)
```
